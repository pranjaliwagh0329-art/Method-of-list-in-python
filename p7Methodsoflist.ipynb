### Loop list

```python
A=["riya","siya","mira"]
for i in A:
    print(i)
```

```python
A=["riya","siya","mira"]
for i in range(len(A)):
    print(A[i])
```

### 12) list comprehension

```python
fruits=["Cherry","apple","banana","mango","kiwi"]
Nlist=[]
for x in fruits:
  if "a" in x:
    Nlist.append(x)
print(Nlist)
 
```

```python
fruits=["Cherry","apple","banana","mango","kiwi"]
Nlist=[]
for x in fruits:
  if "r" in x:
    Nlist.append(x)
print(Nlist)
```
